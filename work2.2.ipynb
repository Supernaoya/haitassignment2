```python
#'Age','Cabin'に欠損があるサンプルのみ削除する
titanic2_df=titanic_df.dropna(subset=['Age','Cabin'])
titanic2_df
```

```python
#'Age','Cabin'に欠損があるサンプルのみ削除する
test2_df=test_df.dropna(subset=['Age','Cabin'])
test2_df
```

```python
# Embarkedと整数を対応させる辞書を生成
Embarked_mapping = {'S':1, 'Q':2, 'B':3, 'C':4}
```

```python
# 平均値補完
titanic3_df=titanic2_df.fillna(titanic2_df.mean(0))
titanic3_df
```

```python
# マッピングの実行
titanic3_df['Embarked'] = titanic_df['Embarked'].map(Embarked_mapping)
titanic3_df
```

```python
# pandasでone-hotエンコーディングを使い、ダミー変数を作成
pd.get_dummies(titanic3_df['Embarked'])
```

```python
# trainデータを用意
# DataFrameの.valuesメソッドでarrayに変換できます
X = titanic3_df.loc[:, ['Age']].values
X
```

```python
# testデータを用意
# DataFrameの.valuesメソッドでarrayに変換できます
y = titanic3_df.loc[:, ['Survived']].values
y
```

```python
# 重回帰の線形回帰を実行
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# 重回帰線形モデルの学習
# trainデータのみを用いていることに注意
lr.fit(X, y)
```

```python
# 定数項を出力
lr.intercept_
```

```python
# 偏回帰係数を出力
# [Age, 'Survived']
lr.coef_
```

```python
from sklearn.model_selection import learning_curve
lc = learning_curve(lr, X, y)
```

```python
# 単回帰モデルを学習
lr.fit(X, y)
```

```python
# 散布図に近似直線を重ねてみる
plt.scatter(X, y)                       # 散布図を表示
plt.plot(X, lr.predict(X), color='red') # 回帰直線を表示
plt.title('年齢と生存')     # グラフのタイトルを表示
plt.xlabel('Age')               # x軸のラベルを表示
plt.ylabel('Survived')                  # y軸のラベルを表示
plt.show()                              # 上記の内容でグラフを表示
```
